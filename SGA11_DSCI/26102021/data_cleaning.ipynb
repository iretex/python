# Operations on Null Values

```python
df=pd.read_csv('../Customer_loan_repayment_problem/data/datasets_137197_325031_test_Y3wMUE5_7gLdaTN.csv')
df.head()
```

```python
df.dropna(axis=1)
```

```python
# total missing observations
df.isnull().sum()
# bool_ex = [True, True,True,False]
# sum(bool_ex)
```

```python
len(df)*0.05
```

```python
# Dropping columns with at least 30% null values 
list(df.columns[df.isnull().sum() > len(df)*0.05 ])
```

```python
df.drop(df.columns[df.isnull().sum() > len(df)*0.05 ],axis=1, inplace=True)
```
