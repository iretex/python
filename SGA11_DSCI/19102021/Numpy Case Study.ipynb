### Extract the last element from each row after the header row

````python
list1 = [1,3,4,7,9]
print(list1)
type(list1[-1])
````

````python
# wine[1:]
# list comprehension
quality = [float(i[-1]) for i in wine[1:]]
# wine[0][-1]
# wine[0,-1]
````

````python
# using a for-loop

# quality = list()

# print(quality)

quality = [] #define an empty list
print("Initial list :",quality)

# iterate through your wine list and append the last element (index -1 position) to the initialized empty list
for w in wine[1:]:
    quality.append(float(w[-1]))

print("final list :", len(quality))

````

````python
# divide the sum of all elements in the quality field by tthe total number of elements in the quality field 
avg_quality = sum(quality)/len(quality)

avg_quality
````
